# Problem Statement

SIB(Small Industries Bank) loans money to companies in exchange for the promise of repayment.  Some will default on the loans, being unable to repay them for some reason. The bank maintains insurance to reduce their risk of loss in the event of default. The insured amount may cover all or just some part of the loan amount.  SIB wants to predict which companies will default on their loans based on their financial information. Using the dataset provided, I have to build a machine learning model predicting which companies will default on their loans and explain how different features impact the predictions.

# Functions

```python
# Function for countplot
def draw_countplot(df, x, title, xlabel = None, ylabel = 'Number of Observations', hue = 'default_status', loc = 'upper right', annot = True):
    fig, ax = plt.subplots(figsize=(18, 8))
    sns.set_style("white")
    sns.countplot(x = x, hue = hue, data = df, dodge = True, palette = ['red', 'blue'])
    ax.set_title(title, fontsize = 20)
    ax.set_ylabel(ylabel, fontsize = 20)
    ax.set_xlabel(xlabel, fontsize = 20)
    legend = plt.legend(loc = loc)
    legend.get_texts()[0].set_text('Have not Defaulted')
    legend.get_texts()[1].set_text('Defaulted')
    
    if annot == True:
        for p in ax.patches:
            ax.annotate(format(p.get_height()), (p.get_x() + p.get_width() / 2, p.get_height()), 
                        ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points', size = 10)
```

```python
# GridSearch Function
def pipeline_gridsearch(pipe, pipe_params, X_train, y_train):
    gs = GridSearchCV(pipe, param_grid=pipe_params, cv=5, n_jobs = -1, scoring = 'accuracy')
    gs.fit(X_train, y_train)
    print('Best Accuracy: ' + str(gs.best_score_)) # best_score_ is gridsearch's function
    print("Best Parameters: " + str(gs.best_params_))# best_params_ is gridsearch's function
```

```python
# Testing the Model
def fit_and_test(best_feature, best_model, X_train, X_test, y_train, y_test, full_list, model):
    X_train_featured = best_feature.fit_transform(X_train)
    X_test_featured = best_feature.transform(X_test)
    best_model.fit(X_train_featured, y_train)
    train_score_val = cross_val_score(best_model, X_train, y_train, cv=5)
    train_score = best_model.score(X_train_featured, y_train)
    test_score = best_model.score(X_test_featured, y_test)
    
    
    predict = best_model.predict(X_test_featured)
    report = classification_report(y_test, best_model.predict(X_test_featured))
    con_matrix = confusion_matrix(y_test, predict)
    tn, fp, fn, tp = confusion_matrix(y_test, predict).ravel()
    if ('SVM' not in model):
        probabilities = best_model.predict_proba(X_test_featured)
    f1 = f1_score(y_test, predict)
    precision = precision_score(y_test, predict)
    recall = recall_score(y_test, predict)
    
    
    #print('{} features were selected'.format(len(selected_features)))
    print('Cross Validated Mean Train Score is: {}'.format(train_score_val.mean()))
    print('Train Accuracy is: {}'.format(train_score))
    print('Test Accuracy is: {}'.format(test_score))
    print('f1 Score is: {}'.format(f1))
    print('Precision is: {}'.format(precision))
    print('Recall is: {}'.format(recall))
    
    # Append Scores into the list
    if model not in full_list[0]:
        full_list[0].append(model)
        full_list[1].append(train_score)
        full_list[2].append(test_score)
        full_list[3].append(f1)
        full_list[4].append(precision)
        full_list[5].append(recall)
    

    return (predict, con_matrix, report, tn, fp, fn, tp)
```

```python
# Evaluation
def evaluation(best_feature, best_model, X_train, X_test, y_train, y_test):
    X_train_featured = best_feature.fit_transform(X_train)
    X_test_featured = best_feature.transform(X_test)
    best_model.fit(X_train_featured, y_train)
    train_score_val = cross_val_score(best_model, X_train, y_train, cv=5)
    train_score = best_model.score(X_train_featured, y_train)
    test_score = best_model.score(X_test_featured, y_test)
    
    
    predict = best_model.predict(X_test_featured)
    report = classification_report(y_test, best_model.predict(X_test_featured))
    con_matrix = confusion_matrix(y_test, predict)
    tn, fp, fn, tp = confusion_matrix(y_test, predict).ravel()
    probabilities = best_model.predict_proba(X_test_featured)
    f1 = f1_score(y_test, predict)
    precision = precision_score(y_test, predict)
    recall = recall_score(y_test, predict)
    
    # Get numerical feature importances
    importances = list(best_model.feature_importances_)# List of tuples with variable and importance
    feature_importances = [(feature, round(importance, 2)) for feature, importance in zip(X_train.columns, importances)]# Sort the feature importances by most important first
    feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)# Print out the feature and importances 
    
    
    #print('{} features were selected'.format(len(selected_features)))
    print('Cross Validated Mean Train Score is: {}'.format(train_score_val.mean()))
    print('Train Accuracy is: {}'.format(train_score))
    print('Test Accuracy is: {}'.format(test_score))
    print('f1 Score is: {}'.format(f1))
    print('Precision is: {}'.format(precision))
    print('Recall is: {}'.format(recall))

    return (predict, con_matrix, report, tn, fp, fn, tp, probabilities, feature_importances)
```

```python
# Define function to calculate sensitivity. (True positive rate.)
def TPR(df, true_col, pred_prob_col, threshold):
    true_positive = df[(df[true_col] == 1) & (df[pred_prob_col] >= threshold)].shape[0]
    false_negative = df[(df[true_col] == 1) & (df[pred_prob_col] < threshold)].shape[0]
    return true_positive / (true_positive + false_negative)
    

# Define function to calculate 1 - specificity. (False positive rate.)
def FPR(df, true_col, pred_prob_col, threshold):
    true_negative = df[(df[true_col] == 0) & (df[pred_prob_col] <= threshold)].shape[0]
    false_positive = df[(df[true_col] == 0) & (df[pred_prob_col] > threshold)].shape[0]
    return 1 - (true_negative / (true_negative + false_positive))
```

## Data Cleaning

```python
# Change Null Value to 'Others'
df.loc[(df['industry'].isnull()), 'industry'] = 'Others'
```

```python
# Change date to datetime format (will be convert to int to apply into model later)
df['request_date'] = df['request_date'].apply(lambda x: datetime.strptime(x, '%d-%b-%y'))
```

```python
df.head()
```

```python
# Change the value of other_loans from (Y & N) to (1 & 0) respectively
df.loc[(df['other_loans'] == 'N'), 'other_loans'] = 0
df.loc[(df['other_loans'] == 'Y'), 'other_loans'] = 1
```

```python
# Change the labels from business_new from New & Existing to 1 & 0 respectively
df.loc[(df['business_new'] == 'New'), 'business_new'] = 1
df.loc[(df['business_new'] == 'Existing'), 'business_new'] = 0
```

```python
# Remove '$' and ',' from loan_amount and insured_amount
df['loan_amount'] = df['loan_amount'].apply(lambda x: x.replace("$", "").replace(",",''))
df['insured_amount'] = df['insured_amount'].apply(lambda x: x.replace("$", "").replace(",",''))
```

```python
# Convert data type of loan_amount and insured_amount to float, business_new and other_loans to int
df['loan_amount'] = pd.to_numeric(df['loan_amount'])
df['insured_amount'] = pd.to_numeric(df['insured_amount'])
df['business_new'] = pd.to_numeric(df['business_new'])
df['other_loans'] = pd.to_numeric(df['other_loans'])
```

## SMOTE
SMOTE stands for Synthetic Minority Oversampling Technique. This is a statistical technique for increasing the number of cases in your dataset in a balanced way. The module works by generating new instances from existing minority cases that you supply as input. This implementation of SMOTE does not change the number of majority cases.

More on SMOTE can be found [here](https://docs.microsoft.com/en-us/azure/machine-learning/studio-module-reference/smote)

## Create Pipeline for model gridsearch

```python
# Creating Pipeline
# LogisticRgression
pipe_lr = Pipeline([
    ('ss', StandardScaler()),
    ('lr', LogisticRegression())
])

# KNN
pipe_knn = Pipeline([
    ('ss', StandardScaler()),
    ('knn', KNeighborsClassifier())
])

# Decision Trees
pipe_dt = Pipeline([
    ('ss', StandardScaler()),
    ('dt', DecisionTreeClassifier())
])

# Random Forest
pipe_rf = Pipeline([
    ('ss', StandardScaler()),
    ('rf', RandomForestClassifier())
])


# AdaBoost
pipe_ada = Pipeline([
    ('ss', StandardScaler()),
    ('ada', AdaBoostClassifier())
])

# Gradient Boost
pipe_gb = Pipeline([
    ('ss', StandardScaler()),
    ('gb', GradientBoostingClassifier())
])

# SVM
pipe_svm = Pipeline([
    ('ss', StandardScaler()),
    ('svm', SVC())
])
```

```python
# Probability Visualisation
pred_proba = [i[1] for i in probabilities]

pred_df = pd.DataFrame({'true_values': y_test_sm,
                        'pred_probs':pred_proba})
```

```python
# Create figure.
plt.figure(figsize = (10,7))

# Create two histograms of observations.
plt.hist(pred_df[pred_df['true_values'] == 0]['pred_probs'],
         bins=25,
         color='red',
         alpha = 0.6,
         label='Outcome = No Default')
plt.hist(pred_df[pred_df['true_values'] == 1]['pred_probs'],
         bins=25,
         color='blue',
         alpha = 0.6,
         label='Outcome = Default')

# Add vertical line at P(Outcome = 1) = 0.5.
plt.vlines(x=0.5,
           ymin = 0,
           ymax = 200,
           color='black',
           linewidth = 3,
           linestyle = '--')

# Label axes.
plt.title('Distribution of P(Outcome = Default)', fontsize=22)
plt.ylabel('Frequency', fontsize=18)
plt.xlabel('Predicted Probability that Outcome = Default', fontsize=18)

# Create legend.
plt.legend(fontsize=20);
```

The above shows the distribution of the predicted probabilities. Colored in blue are the outcomes classified Defaulted and red classified as No Default. 

In a perfect model where every classification is correct, you will see that all the red will be to the left of the white line while all the blues will be to the right. 

We can see that there are misclassifications, the blues that are at the left of the line and red that are at the right. From the above, we can see that the ratio of misclassifications are low compared to the observations where the correct classifications are made.

### ROC curve

```python
# Visualizing the ROC Curve
# Create figure.
plt.figure(figsize = (10,7))

# Create threshold values. (Dashed red line in image.)
thresholds = np.linspace(0, 1, 200)


# Calculate sensitivity & 1-specificity for each threshold between 0 and 1.
tpr_values = [TPR(pred_df, 'true_values', 'pred_probs', prob) for prob in thresholds]
fpr_values = [FPR(pred_df, 'true_values', 'pred_probs', prob) for prob in thresholds]

# Plot ROC curve.
plt.plot(fpr_values, # False Positive Rate on X-axis
         tpr_values, # True Positive Rate on Y-axis
         label='ROC Curve')

# Plot baseline. (Perfect overlap between the two populations.)
plt.plot(np.linspace(0, 1, 200),
         np.linspace(0, 1, 200),
         label='baseline',
         linestyle='--')

# Label axes.
plt.title(f'ROC Curve with AUC = {round(roc_auc_score(pred_df["true_values"], pred_df["pred_probs"]),3)}', fontsize=22)
plt.ylabel('Sensitivity', fontsize=18)
plt.xlabel('1 - Specificity', fontsize=18)

# Create legend.
plt.legend(fontsize=16);
```

From the above we can see that our ROC AUC is 0.983. we can see that our model is doing phenominally well.

- term
- loan_amount
- other_loans

This can be expected as the duration of the loan and the loan amount has direct impact to whether the business would be able to repay the loan or not. Other loans would show whether the business is currenly in debt and how much extra borrowing would affect the business.

Something that I noticed when I dived deeper into the important features:
- Industry and State columns are not deemed important based on noticeable differences in its default rates. Instead, the features are important to the model because those states and industries have the highest numbers of loan borrowers compared to the others.

Possible Bottleneck in the model:
- Because of the way that the model indentified the important features, it might not be so accurate in its prediction as the pattern identified for the states and industries is not based on the default rates (Correct). But instead it is based on highest number of observations (Wrong). This might cause issues when making predictions in different states.

Future work to improve model:
- Increase the number of observations in each state and industry such that the number of observations are more or less equal. This is to prevent the model from deciding that a feature is important based on the number of observations. Instead, it will force the model to identify other patterns, which hopefully would be the difference in default rates.

## Clean up test data
We have to clean up the test dataset to make it consistent with our training data in order to run the model and get accurate predictions

```python
# Change date to datetime format
test['request_date'] = test['request_date'].apply(lambda x: datetime.strptime(x, '%d-%b-%y'))
# Convert Month to numerical to be able to use in model
test['request_date'] = test['request_date'].map(dt.datetime.toordinal)

# Change the value of other_loans from (Y & N) to (1 & 0) respectively
test.loc[(test['other_loans'] == 'N'), 'other_loans'] = 0
test.loc[(test['other_loans'] == 'Y'), 'other_loans'] = 1

# Change the labels from business_new from New & Existing to 1 & 0 respectively
test.loc[(test['business_new'] == 'New'), 'business_new'] = 1
test.loc[(test['business_new'] == 'Existing'), 'business_new'] = 0

# Remove '$' and ',' from loan_amount and insured_amount
test['loan_amount'] = test['loan_amount'].apply(lambda x: x.replace("$", "").replace(",",''))
test['insured_amount'] = test['insured_amount'].apply(lambda x: x.replace("$", "").replace(",",''))

# Convert data type of loan_amount and insured_amount to float, business_new and other_loans to int
test['loan_amount'] = pd.to_numeric(test['loan_amount'])
test['insured_amount'] = pd.to_numeric(test['insured_amount'])
test['business_new'] = pd.to_numeric(test['business_new'])
test['other_loans'] = pd.to_numeric(test['other_loans'])
```
